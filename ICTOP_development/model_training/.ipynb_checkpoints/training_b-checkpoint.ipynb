```python
def save_model(bst: xgb.Booster, modelname: str):
    print("modelname: " + model_name)
    # save model
    bst.save_model(model_name)
```

```python
def train(dtrain:xgb.DMatrix, dtest: xgb.DMatrix, num_round: int) -> xgb.Booster:
    # get train test split 
    # dtrain = xgb.DMatrix( X_train, label=y_train)  # so ist mein kompletter Datensatz da drin
    # dtest = xgb.DMatrix(X_test, label=y_test)

    num_classes = len(encoder_organ.classes_)
    print("Number of classes:", num_classes)

    # booster parameters
    param = {
        "max_depth": 6,
        "eta": 1,
        "objective": "multi:softmax",
        "num_class": num_classes,
        "nthread": 4,
        "eval_metric": "auc",
        "booster": "gbtree"  # Specify tree booster
    }
    
    # param = {'max_depth': 2, 'eta': 1, 'objective': 'binary:logistic'}
    # param['nthread'] = 4
    # param['eval_metric'] = 'auc'
    
    evallist = [(dtrain, "train"), (dtest, "eval")]
    bst = xgb.train(param, dtrain, num_round, evallist)

    return bst
```

```python
def load_model(modelname: str) -> xgb.Booster:
    bst = xgb.Booster()
    bst.load_model(modelname)
    return bst
```

```python
def feature_importance_scores(bst: xgb.Booster):
    # Obtain feature importance scores
    feature_importance = bst.get_score(importance_type="weight")

    # Rank features based on importance scores
    sorted_features = sorted(
        feature_importance.items(), key=lambda x: x[1], reverse=True
    )

    # Extract top predictive features
    top_features = [
        feature[0] for feature in sorted_features[:10]
    ]  # Adjust 10 to the desired number of top features

    # Visualize feature importance
    plt.figure(figsize=(10, 6))
    plt.bar(
        range(len(sorted_features)),
        [feature[1] for feature in sorted_features],
        tick_label=[feature[0] for feature in sorted_features],
    )
    plt.xlabel("Feature")
    plt.ylabel("Importance Score")
    plt.title("Feature Importance")
    plt.xticks(rotation=90)
    plt.show()

    # Print top predictive features
    print("Top Predictive Features:")
    for feature in top_features:
        print(feature)

    # Feature importances
    feature_importance = bst.get_score(importance_type="weight")
    
    # Rank features based on importance scores
    sorted_features = sorted(
        feature_importance.items(), key=lambda x: x[1], reverse=True
    )
    
    # Visualize feature importance
    plt.figure(figsize=(10, 6))
    plt.bar(
        range(len(sorted_features)),
        [feature[1] for feature in sorted_features],
        tick_label=[feature[0] for feature in sorted_features],
    )
    plt.xlabel("Feature")
    plt.ylabel("Importance Score")
    plt.title("Feature Importance")
    plt.xticks(rotation=90)
    plt.show()
    
    
    # Begrenze auf die Top 40 Merkmale
    top_features = sorted_features[:90]
    
    # Visualisiere die Merkmalswichtigkeit für die Top 40 Merkmale
    plt.figure(figsize=(10, 6))
    plt.bar(
        range(len(top_features)),
        [feature[1] for feature in top_features],
        tick_label=[feature[0] for feature in top_features],
    )
    plt.xlabel("Merkmal")
    plt.ylabel("Wichtigkeit Score")
    plt.title("Top 40 Merkmale nach Wichtigkeit")
    plt.xticks(rotation=90)
    plt.show()

    xgb.plot_importance(bst, max_num_features=25)
    plt.title('xgboost.plot_importance(model) - Top 25 Merkmale')
    plt.show()

    xgb.plot_importance(bst, importance_type="cover", max_num_features=25)
    plt.title('xgboost.plot_importance(model, importance_type="cover") - Top 25 Merkmale')
    plt.show()

    xgb.plot_importance(bst, importance_type="gain", max_num_features=25)
    plt.title('xgboost.plot_importance(model, importance_type="gain") - Top 25 Merkmale')
    plt.show()
    
```

```python
def prediction(bst: xgb.Booster) -> pd.DataFrame: # gets model and returns classification report, y_pred_encoded oder adata mit prediction
    y_pred_encoded = bst.predict(dtest)

    accuracy = accuracy_score(y_test, y_pred_encoded)
    print(f"Accuracy: {accuracy}")
    
    y_test_decoded = encoder_organ.inverse_transform(
        y_test
    )  # todo hier andere variablen namen
    y_pred_encoded_series = pd.Series(y_pred_encoded)

    # Map the encoded labels to decoded labels using the label table DataFrame
    y_pred = y_pred_encoded_series.map(
        dict(zip(label_table["Encoded_Label"], label_table["Decoded_Label"]))
    )

    class_report = metrics.classification_report(y_test, y_pred_encoded)

    print("Classification Report for celltype: ", celltype)
    print(class_report)

    with open(report_name, "a") as f:
        print("Classification Report for celltype:", celltype, file=f)
        print(class_report, file=f)

    # Assuming your DataFrame is named df
    # Get value counts of 'Actual' and 'Predicted' columns
    result_df = pd.DataFrame({"Actual": y_test, "Predicted": y_pred_encoded})

    actual_counts = result_df["Actual"].value_counts()
    predicted_counts = result_df["Predicted"].value_counts()

    # Combine the counts into one DataFrame
    combined_counts = pd.concat([actual_counts, predicted_counts], axis=1)
    combined_counts.columns = ["Actual", "Predicted"]
    combined_counts = combined_counts.fillna(0)  # Fill NaN values with 0 if any

    # Plot the distributions
    combined_counts.plot(kind="bar", figsize=(12, 6))
    plt.title("Distribution of Actual and Predicted Categories for " + celltype)
    plt.xlabel("Category")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.legend(["Actual", "Predicted"])
    plt.show()

    return y_pred_encoded
```

# Parameter tuning : 
https://blog.cambridgespark.com/hyperparameter-tuning-in-xgboost-4ff9100a3b2f

```python
X_train, X_test, y_train, y_test = ttsplit(celltype) # returns X_train,...
# X and y contain whole dataset, not splitted to a specific celltype? 
X = adata.to_df()
y = adata.obs["Organ"]

dtrain = xgb.DMatrix(X_train, label=y_train)  # so ist mein kompletter Datensatz da drin
dtest = xgb.DMatrix(X_test, label=y_test)
```

```python
num_classes = len(encoder_organ.classes_)
print("Number of classes:", num_classes)
class_names = encoder_organ.classes_
model_number = "0005"
report_name = f"{model_number}_classification_report.txt"
#num_round = 21
# num_boost_round= 21
```

```python
params = {
    # Parameters that we are going to tune.
    'max_depth':6,
    'min_child_weight': 1,
    'eta':.3,
    'subsample': 1,
    'colsample_bytree': 1,
    "num_class": num_classes,
    # Other parameters
    "objective": "multi:softprob",
    "eval_metric": ['merror','mlogloss'], 
    'device': 'cuda'
}
```

### Parameters num_boost_round and early_stopping_rounds

```python
# train model with that
evals_result = {}

model = xgb.train(
    params,
    dtrain,
    num_boost_round=999,
    evals=[(dtrain, "train"), (dtest, "eval")],
    early_stopping_rounds=10,
    evals_result = evals_result
)
print("Best merror: {:.2f} in {} rounds".format(model.best_score, model.best_iteration+1))
```

```python
# Anzahl der Epochen
epochs = len(evals_result['train']['mlogloss'])
x_axis = range(0, epochs)

# Log-Loss für Train und Eval plotten
fig, ax = plt.subplots()
ax.plot(x_axis, evals_result['train']['mlogloss'], label='Train')
ax.plot(x_axis, evals_result['eval']['mlogloss'], label='Eval')
ax.legend()
plt.ylabel('Log Loss')
plt.xlabel('iteration')
plt.title('Log Loss')
plt.show()
```

```python
# xgboost 'merror' plot
fig, ax = plt.subplots(figsize=(9,5))
ax.plot(x_axis, evals_result['train']['merror'], label='Train')
ax.plot(x_axis, evals_result['eval']['merror'], label='Test')
ax.legend()
plt.ylabel('merror')
plt.xlabel('iteration')
plt.title('XGBoost merror')
plt.show()
```

```python
# saving that 
evals_result = {}

num_boost_round = model.best_iteration + 1
best_model = xgb.train(
    params,
    dtrain,
    num_boost_round=num_boost_round,
    evals=[(dtest, "Test")],
    evals_result = evals_result
)
```
